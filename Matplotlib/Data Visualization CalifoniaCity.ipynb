```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cities =pd.read_csv("./california_cities.csv")
cities.head(5)
```

```python
lat,lon = cities["latd"],cities["longd"]
population,area =cities.population_total,cities.area_total_km2
```

```python
plt.figure(figsize=(10,8))
plt.style.use('seaborn-v0_8')
plt.scatter(lon,lat,c=np.log10(population),cmap='viridis',s=area,linewidths=0,alpha=0.6)

plt.axis('equal')
plt.xlabel('longtitude')
plt.ylabel('latitude')
plt.colorbar(label="Log$_{10}$(population)")
plt.clim(0,7)
plt.title('Califonia: Population and area')

area_range =[50,100,300,500]

for area in area_range:
    plt.scatter([],[],s=area,c='k',alpha=0.4,
               label=str(area)+" km$^2$")
    plt.legend(scatterpoints=1,labelspacing=1,title='city area')
```
