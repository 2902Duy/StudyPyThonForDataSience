# 2. Most common types of Matplotlib plots
- **line**
- scatter
- bar
- hist
- `subplots()`

```python
x= np.linspace(0,10,100)
x[:5]
```

### Line

```python

fig, ax =plt.subplots()
ax.plot(x,x**3)
```

### Scatter

```python
#plot API
plt.scatter(x,x**3)
```

```python
#OO API
rd = np.random
x= rd.random(100)
y = rd.random(100)

sizes =500*rd.rand(100)
colors =rd.randn(100)
fig,ax =plt.subplots()
img1= ax.scatter(x,y,s=sizes,c=colors,cmap='viridis',alpha=0.3)
fig.colorbar(img1)
```

### Bar
- vertical
- horizontal

```python
soft_drink_price ={"CoCa":10,
                  "Pepsi":12,
                  "333":14}
fig, ax = plt.subplots()
ax.set(title="Vertical bar")
ax.bar(soft_drink_price.keys(),soft_drink_price.values())
```

```python
fig, ax = plt.subplots()
ax.barh(list(soft_drink_price.keys()),list(soft_drink_price.values()))
ax.set(title="Horizontal bar")
```

### Histogram

```python
student_high = np.random.normal(179,10,200)
student_high[:5]
```

```python
# plot API
plt.hist(student_high,bins= 20)
```

### Sublots

```python
fig, ((ax1,ax2)) = plt.subplots(nrows=1,ncols=2,figsize=(10,5))
ax1.plot(x,x*2)
ax1.plot(x,x*3)
ax2.scatter(np.random.rand(10),np.random.rand(10))
```
