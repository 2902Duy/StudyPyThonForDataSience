```python
import os
import time
```

```python
from webdriver_manager.chrome import ChromeDriverManager
from tqdm import tqdm
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
chrome_option= webdriver.ChromeOptions()
#chrome_option.add_argument("strart_maximized")
chrome_option.add_argument("--headless=new")
chrome_option.add_argument("--disable-dev-shm-usage")
chrome_option.add_argument("--no-sandbox")

driver = webdriver.Chrome(
    service=Service(ChromeDriverManager().install()),
    options=chrome_option
)

#url='https://www.python.org/'
#driver.get(url)
```

```python
root_dir='./vn_news'
os.makedirs(root_dir,exist_ok=True)
n_pages=6
article_id=0
```

```python
for page_idx in range(n_pages):
    main_url=f'https://vietnamnet.vn/thoi-su-page{page_idx}'
    driver.get(main_url)
    new_xpath='//div[@class="topStory-15nd"]/div/div[1]/a'
    news_tags=driver.find_elements(
        By.XPATH,
        new_xpath
    )
    new_page_urls =[
        news_tag.get_attribute('href') \
        for news_tag in news_tags
    ]
    print(new_page_urls)
    
```

```python
for news_page_url in new_page_urls:
    driver.get(news_page_url)
    time.sleep(1)  # Tùy chỉnh thời gian chờ theo tốc độ tải trang
    print(driver)
    
    try:
        main_content_tag = driver.find_element(By.XPATH, '//div[@class="content-detail content-mobile-change"]')
        
        
    except Exception as e:
        print(f"Main content not found for URL: {news_page_url}, Error: {e}")
        continue
    print(main_content_tag)
    try:
        video_content_tag = driver.find_element(By.XPATH, '//div[@class="video-detail"]')
    except Exception as e:
        #print(f"Video content not found for URL: {news_page_url}, Error: {e}")
        pass

    try:
        title = main_content_tag.find_element(By.TAG_NAME, 'h1').text.strip()
        print(f"Title: {title}")
    except Exception as e:
        print(f"Title not found, Error: {e}")
        title = ''

    try:
        abstract = main_content_tag.find_element(By.TAG_NAME, 'h2').text.strip()
        print(f"Abstract: {abstract}")
    except Exception as e:
        print(f"Abstract not found, Error: {e}")
        abstract = ''

    try:
        author = main_content_tag.find_element(By.XPATH, 'span[@class="name"]').text.strip()
    except Exception as e:
        print(f"Author not found, Error: {e}")
        author = ''

    try:
        paragraph_tags = main_content_tag.find_elements(By.XPATH, 'div[@class="maincontent main-content"]/p')
        paragraph_lst = [paragraph.text.strip() for paragraph in paragraph_tags]
        print(f"Paragraphs: {paragraph_lst}")
    except Exception as e:
        print(f"Paragraphs not found, Error: {e}")
        paragraph_lst = []

    # Lưu nội dung
    final_content = '\n\n'.join([title, abstract, author] + paragraph_lst)
    article_name = f"article_{article_id:05d}.txt"
    article_savepath = os.path.join(root_dir, article_name)

    with open(article_savepath, 'w', encoding='utf-8') as f:
        f.write(final_content)

    print(f"Saved article: {article_savepath}")
    article_id += 1

driver.quit()
```
