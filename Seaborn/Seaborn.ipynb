```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
print(sns.get_dataset_names())
```

```python
tip_df = sns.load_dataset('tips')
```

```python
tip_df.head(5)
```

# histogram

```python
sns.histplot(tip_df.total_bill)
sns.
```

```python
tip_df.total_bill.value_counts().sort_values(ascending=False)
```

# KDE

```python
sns.kdeplot(tip_df.total_bill)
```

```python
sns.displot(data=tip_df,x='total_bill',col='time',kde=True)
```

```python
sns.barplot(data=tip_df,x='smoker',y='tip')
```

```python
sns.countplot(data=tip_df,x='sex')
```

```python
sns.boxplot(data=tip_df,x='day',y='total_bill',hue='sex')
plt.legend(loc=0)
```

```python
tip_df.head(5)
```

```python
tip_fig= sns.FacetGrid(data=tip_df,row='time',col='smoker')
tip_fig.map(sns.scatterplot,'total_bill','tip')
```
