```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cereal_df = pd.read_csv('./cereal.csv')
```

```python
cereal_df.head()
```

```python
fields=['shelf','weight','cups','rating']
cereal_df_new =cereal_df.drop(fields,axis=1)
cereal_df_new.head()
```

```python
numeric_df = cereal_df_new.select_dtypes(include=[float, int])

# Calculate the correlation matrix
cereal_corr = numeric_df.corr()
cereal_corr
```

```python
ones_corr= np.ones_like(cereal_corr,dtype=bool)
ones_corr
```

```python
mask= np.triu(ones_corr)
mask
```

```python
sns.heatmap(data=cereal_corr,mask=mask)
```

```python
a_mask =mask[1:,:-1]
a_mask
```

```python
a_cereal_corr= cereal_corr.iloc[1:,:-1]
```

```python
sns.heatmap(data=a_cereal_corr,mask=a_mask,cmap='Blues',annot=True)
```
